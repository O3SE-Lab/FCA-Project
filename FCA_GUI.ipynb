```python
import pandas as pd
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QDesktopWidget, QGridLayout, QPushButton, QFileDialog, QLabel, QTextBrowser, QTextEdit
```

```python
class FCAwithGUI(QWidget):

    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        # 기본 창 설정
        self.setWindowTitle('FCA with GUI')
        self.resize(800,900)
        self.center()

        # 그리드 레이아웃을 사용한 위젯 배치 및 설정
        self.grid = QGridLayout()
        self.setLayout(self.grid)

        # InputContext - 라벨 생성 및 설정
        self.ICtitleL = QLabel('<< Input Context >>', self)  # IC 타이틀 라벨
        self.ICfptitleL = QLabel('[선택 파일]', self) # 선택 파일 타이틀 라벨
        self.ICfilePathL = QLabel('', self)  # 선택한 파일 경로를 보여줄 라벨
        self.ICresultTitleL = QLabel('[Input Context 실행결과]', self)  # IC 실행결과 타이틀 라벨

        self.grid.addWidget(self.ICtitleL, 0, 0)
        self.grid.addWidget(self.ICfptitleL, 2, 0)
        self.grid.addWidget(self.ICfilePathL, 3, 0)
        self.grid.addWidget(self.ICresultTitleL, 4, 0)

        # InputContext - 파일 오픈 버튼 생성 및 설정
        self.ICfileOpenB = QPushButton('Context 입력(Open CSV File)', self)
        self.grid.addWidget(self.ICfileOpenB, 1, 0)
        self.ICfileOpenB.clicked.connect(self.csv_open)

        # InputContext - 실행 버튼 생성 및 설정
        self.ICexcuteB = QPushButton('실행', self)
        self.grid.addWidget(self.ICexcuteB, 1, 1)
        self.ICexcuteB.clicked.connect(self.IC_clear_text)
        self.ICexcuteB.clicked.connect(lambda: self.InputContext(self.ICfilePathL.text()))

        # InputContext - 실행 결과 출력 화면 생성 및 설정
        self.ICtb = QTextBrowser()
        self.grid.addWidget(self.ICtb, 5, 0)
        
        # CACO - 라벨 생성 및 설정
        self.ntitleL = QLabel("",self) # 빈 타이틀 라벨
        self.CACOtitleL = QLabel('<< CA & CO >>', self)  # CACO 타이틀 라벨
        self.CAtitleL = QLabel('[CA]', self) # CA 타이틀 라벨
        self.CAresultTitleL = QLabel('[CA 실행결과]', self)  # CA 실행결과 타이틀 라벨
        self.COtitleL = QLabel('[CO]', self) # CO 타이틀 라벨
        self.COresultTitleL = QLabel('[CO 실행결과]', self)  # CO 실행결과 타이틀 라벨
        
        self.grid.addWidget(self.ntitleL, 5, 0)
        self.grid.addWidget(self.CACOtitleL, 6, 0)
        self.grid.addWidget(self.CAtitleL, 7, 0)
        self.grid.addWidget(self.CAresultTitleL, 9, 0)
        self.grid.addWidget(self.COtitleL, 11, 0)
        self.grid.addWidget(self.COresultTitleL, 13, 0)
        
        # CA - 실행 버튼 생성 및 설정
        self.CAexcuteB = QPushButton('실행', self)
        self.grid.addWidget(self.CAexcuteB, 8, 1)
        self.CAexcuteB.clicked.connect(self.CA_clear_text)
        self.CAexcuteB.clicked.connect(lambda: self.CA(self.CAte.toPlainText()))
        
        # CA - 입력창 생성 및 설정
        self.CAte = QTextEdit()
        self.grid.addWidget(self.CAte, 8, 0)
        self.CAte.resize(10, 10)
        
        # CA - 결과 출력 화면 생성 및 설정
        self.CAtb = QTextBrowser()
        self.grid.addWidget(self.CAtb, 10, 0)
        
        # CO - 실행 버튼 생성 및 설정
        self.COexcuteB = QPushButton('실행', self)
        self.grid.addWidget(self.COexcuteB, 12, 1)
        self.COexcuteB.clicked.connect(self.CO_clear_text)
        self.COexcuteB.clicked.connect(lambda: self.CO(self.COte.toPlainText()))
        
        # CO - 입력창 생성 및 설정
        self.COte = QTextEdit()
        self.grid.addWidget(self.COte, 12, 0)
        
        # CO - 결과 출력 화면 생성 및 설정
        self.COtb = QTextBrowser()
        self.grid.addWidget(self.COtb, 14, 0)
        
        self.show()

    # Input Context 실행 함수
    def InputContext(self, fPath):

        data = pd.read_csv(fPath)

        # G : 객체들의 집합, M : 속성들의 집합, I : 객체와 속성 사이의 관계 집합 (I ⊆ G × M)
        G = set(data.index)
        M = set(data.columns)
        I = set()

        # Process : Formal Context를 읽어들여서 "적절한 Data구조 K"에 저장". formal context K:=(G, M, I)
        for g in G:
            for m in M:
                if data[m][g] in ['x', 'X']:
                    I.add((g, m))

        K = [G, M, I]
        
        self.ICtb.append(" G ".center(80,'-'))
        self.ICtb.append(str(G))
        self.ICtb.append(" M ".center(80,'-'))
        self.ICtb.append(str(M))
        self.ICtb.append(" I ".center(80,'-'))
        self.ICtb.append(str(I))
        self.ICtb.append(" K ".center(80,'-'))
        self.ICtb.append(str(K))
        
        return K
    
    # CA
    def CA(self, A):
        pass

    # CO
    def CO(self, B):
        pass
    
    # ICtextBrowser clear 함수
    def IC_clear_text(self):
        self.ICtb.clear()
    
    # CAtextBrowser clear 함수
    def CA_clear_text(self):
        self.CAtb.clear()
    
    # COtextBrowser clear 함수
    def CO_clear_text(self):
        self.COtb.clear()
    
    # 창을 화면 가운데에 띄우는 함수
    def center(self):
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())
        
    # csv file open 및 filePath 출력 함수
    def csv_open(self):
        fname = QFileDialog.getOpenFileName(self, 'Open file', './', 'csv(*.csv)')
        self.ICfilePathL.setText(fname[0])
```

```python
app = QApplication(sys.argv)
ex = FCAwithGUI()
app.exec_()
```
